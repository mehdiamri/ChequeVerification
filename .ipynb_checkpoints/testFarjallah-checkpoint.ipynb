```python
# Noise Reduce
def noise_reduce(image):
    return cv2.fastNlMeansDenoisingColored(image,None,10,10,7,21)

# Gray
def to_gray(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Erode
def erode(image):
    kernel = np.ones((5,5),np.uint8)
    return cv2.erode(image, kernel, iterations = 2)

# Detect Countours
def detect_countours(image):
    ret,thresh = cv2.threshold(image,127,255,0)
    contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    return contours, hierarchy

# Delete Lines
def delete_lines(image):
    edges = cv2.Canny(image,30,10,apertureSize = 3)

    lines = cv2.HoughLines(edges,1,np.pi/120,50)
    for rho,theta in lines[0]:
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a*rho
        y0 = b*rho
        x1 = int(x0 + 1600*(-b))
        y1 = int(y0 + 1600*(a))
        x2 = int(x0 - 1600*(-b))
        y2 = int(y0 - 1600*(a))
        cv2.line(image,(x1,y1),(x2,y2),(255, 255, 255),10)
    return image

# Morphologie
def open_mor(image,kernel):
    return cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)
def close_mor(image,kernel):
    return cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
    
def noise_removal(image):
    import numpy as np
    kernel = np.ones((1, 1), np.uint8)
    image = cv2.dilate(image, kernel, iterations=1)
    kernel = np.ones((1, 1), np.uint8)
    image = cv2.erode(image, kernel, iterations=1)
    image = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)
    image = cv2.medianBlur(image, 3)
    return (image)
def thick_font(image):
    import numpy as np
    image = cv2.bitwise_not(image)
    kernel = np.ones((2,2),np.uint8)
    image = cv2.dilate(image, kernel, iterations=1)
    image = cv2.bitwise_not(image)
    return (image)
def Seg(img,i):
    ret,thresh = cv2.threshold(img,150,255,cv2.THRESH_BINARY_INV)
    kernel = np.ones((5,5), np.uint8)
    img_dilation = cv2.dilate(thresh, kernel, iterations=4)

    ctrs, hier = cv2.findContours(img_dilation.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    sorted_ctrs = sorted(ctrs, key=lambda ctr: cv2.boundingRect(ctr)[0])
    for i, ctr in enumerate(sorted_ctrs):
        # Get bounding box
        x, y, w, h = cv2.boundingRect(ctr)
        # Getting ROI
        roi = gray[y:y+h, x:x+w]
        # show ROI
        cv2.rectangle(gray,(x,y),( x + w, y + h ),(0,0,255),2)
    return gray
```

```python
for i in range(0,len(amounts_path)):

    img = cv2.imread(amounts_path[i])
    #gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    #cv2.imwrite("Preprocessing/BWToGray/mage_final"+str(i)+".png",gray)
    #thresh= cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]
    horizontal_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (10, 1))
    detected_lines = cv2.morphologyEx(thresh, cv2.MORPH_OPEN,horizontal_kernel, iterations=2)
    cnts = cv2.findContours(detected_lines, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    cnts = cnts[0] if len(cnts) == 2 else cnts[1]

    for c in cnts:
        cv2.drawContours(img, [c], -1, (0, 0, 0), 2)

    repair_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (1, 6))
    result = 255 - cv2.morphologyEx(255 - img, cv2.MORPH_CLOSE,repair_kernel,iterations=1)

    result = delete_lines(result)
    cv2.imwrite("Preprocessing/edges/mage_final"+str(i)+".png",result)

```
